# Data Cleaning

```python
df.head()
```

### Check for missing values


```python

missing_values_count = df.isnull().sum()
missing_values_count
```

### Check data types


```python

print(df.dtypes)
```

### Sale Prices has values which cannot be turned into integers. so I will replace the values with Nans, then make the values integer


```python

df['SALE PRICE']= pd.to_numeric(df['SALE PRICE'], errors='coerce')


```

```python

df['SALE PRICE'] = df['SALE PRICE'].astype('Int64')
```

### Drop NANs


```python
#sum of Nans
na_counts = df.isna().sum()
na_counts
```

```python
df.shape
```

```python

df_cleaned = df.dropna()
```

```python
df_cleaned.shape
```

### Drop Unnamed: 0 column


```python
df_cleaned = df_cleaned.drop(columns=['Unnamed: 0'])
```

```python
df_cleaned.head()
```

### I want to change borough from number to name to make it easier to use for visualizations


```python

#1 = manhattan
#2 = bronx
#3 = Brooklyn
#4 = Queens
#5= staten island

boro = df_cleaned[df_cleaned['BOROUGH'] == 5]
boro.head()
```

```python
df_cleaned['BOROUGH'] = df_cleaned['BOROUGH'].replace(1, 'Manhattan')
df_cleaned['BOROUGH'] = df_cleaned['BOROUGH'].replace(2, 'Bronx')
df_cleaned['BOROUGH'] = df_cleaned['BOROUGH'].replace(3, 'Brooklyn')
df_cleaned['BOROUGH'] = df_cleaned['BOROUGH'].replace(4, 'Queens')
df_cleaned['BOROUGH'] = df_cleaned['BOROUGH'].replace(5, 'Staten Island')



```

```python
df_cleaned.head()
```

### Export cleaned data to excel


```python

excel_file_path = (r"C:\Users\19178\OneDrive\Documents\Downloads\datasets\nyc-rolling-sales_cleaned.xlsx")
df_cleaned.to_excel(excel_file_path, index=False)
```
